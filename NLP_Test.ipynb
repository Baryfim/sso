```python
from bs4 import BeautifulSoup
import requests
```

```python
def parser(url: str):
    try:
      response = requests.get(url)
      if response.status_code != 200:
        return

      soup = BeautifulSoup(response.text, "html.parser")

      tags_to_search = ['h1', 'h2', 'h3', 'h4', 'h5', 'span', 'p', 'title', 'b', 'strong', 'i', 'em', 'mark', 'small', 'del', 'ins', 'sub', 'sup', 'a', 'li']
      text_tags_list = []

      for tag in tags_to_search:
          text_tags_list.extend([label.get_text(strip=True) for label in soup.find_all(tag)])

      return list(filter(lambda x: x != '', text_tags_list))
    except:
      return
```

```python
# Подключение к cvs файлу с ссылками
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
path = "/content/drive/MyDrive/URL_list.csv"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
import pandas as pd
df = pd.read_csv(path)
```

```python
x_train_data = []
count: int = 0

valid_links_ids = []

for link in df['max(page)']:
    count += 1
    response_data_text = parser(link)

    if response_data_text is not None and len(response_data_text) > 0:
        combined_text = ' '.join(response_data_text)

        if combined_text.strip():
            x_train_vectorized = vectorizer.fit_transform([combined_text])
            x_train_data.append(x_train_vectorized.toarray())

            print(count)

            valid_links_ids.append(count)
```

```python
with open('result.txt', "w+") as f:
  f.write(f"{valid_links_ids}")
```
